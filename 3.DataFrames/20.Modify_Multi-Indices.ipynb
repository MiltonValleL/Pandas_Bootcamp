# MODIFYING MULTI-INDICES

---

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv')
retail.head(3)
```

```python
sale_sum = retail.groupby(['family', 'store_nbr', 'onpromotion'])[['sales']].sum()
sale_sum
```

There are several ways to **modify multi-index DataFrames**:

---

1. `Reset the Index`: Moves the index levels back to DataFrame columns

```python
# Returns into an original DataFrame
sale_sum.reset_index()
```

---

2. `Swap the Index Level`: Changes the hierarchy for the index levels

```python
sale_sum.swaplevel()
```

---

3. `Drop an Index Level`: Drops an index level from the DataFrame entirely

```python
sale_sum.droplevel('family')
```

<span style='color:orange'>
    
`NOTE:` Be careful with this method. You **may lose important information**

## ASSIGNMENT: MULTI-INDEX DATAFRAMES

```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date'])
transactions = transactions.assign(month = transactions.date.dt.month)
transactions.head()
```

```python
grouped_df = (
    transactions.groupby(['store_nbr', 'month'])
    .agg({'transactions':['sum', 'mean', 'max']})
    .sort_values(by=['month', ('transactions', 'sum')], ascending=[True, False])    
)
grouped_df
```

---

NOW: Convert the Multi-Index DataFrame into an `Integer-Based Index`

```python
regular_df = grouped_df.droplevel(level=0, axis=1)
regular_df
```

```python
regular_df = regular_df.reset_index()
regular_df
```
