# ACCESSING DATAFRAME DATA WITH `.iloc[]` and `.loc[]`

---

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Data/retail_2016_2017.csv')
df.iloc[:3, :]
```

```python
df.iloc[:, 1:4]
```

```python
df.loc[:, 'date':'sales']
```

```python
df.loc[:, ['date', 'sales']]
```
