# TRANSFORMING DATAFRAMES

---

>The `.transform()` Method can be used to perform aggregations without reshaping.
>
>   - This is useful for **calculating group-level statistics to perform row-level analysis**

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv')
retail.head(3)
```

```python
retail.groupby('store_nbr')['sales'].sum().round()
```

---

## 1. `.transform()` Method Example

```python
retail.assign(store_sales = (retail
    .groupby('store_nbr')['sales']
    .transform('sum'))).round()
```

## LABORATORY



```python
premier_league = pd.read_excel('Data/premier_league_games_full.xlsx')
premier_league.head(3)
```

```python
performance = premier_league.assign(
    avg_team_goals = premier_league.groupby('HomeTeam')['HomeGoals'].transform('mean'),
    difference = lambda x: x['HomeGoals'] - x['avg_team_goals']
)

performance
```

```python
performance.groupby(['HomeTeam', 'AwayTeam']).agg({'difference':'mean'}).sort_values('difference')
```

```python
performance.query("HomeTeam == 'Arsenal' and AwayTeam == 'Blackpool'")
```

```python
performance.query("HomeTeam == 'Arsenal' and AwayTeam == 'Blackpool'")
```

```python
performance.query("AwayTeam == 'Blackpool'")
```

## ASSIGNMENT: TRANSFORM

```python
transactions = pd.read_csv('Data/transactions.csv')
transactions
```

```python
transactions = transactions.assign(
    target_pct = transactions.transactions / 2500,
    met_target = transactions.transactions / 2500 >=1,
    bonus_payable = (transactions.transactions / 2500 >=1) * 100,
    month = transactions.date.astype('datetime64[ns]').dt.month,
    day_of_week = transactions.date.astype('datetime64[ns]').dt.dayofweek
)
transactions.head()
```

```python
transactions = transactions.assign(
    store_avg_trans = transactions.groupby(['store_nbr', 'day_of_week'])['transactions'].transform('mean'),
    trans_vs_avg = lambda x: x['transactions'] - x['store_avg_trans']
)
transactions
```
