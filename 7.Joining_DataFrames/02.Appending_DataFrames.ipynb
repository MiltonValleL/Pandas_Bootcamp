# APPENDING & JOINING DATAFRAMES

---

>---
>
>There are two ways to combine DataFrames: by **Appending and Joining**
>
>- **Appending** stacks the rows from multiple DataFrames with the same column structure. 
>
>- **Joining** adds related columns from one DataFrame to another, based on common values
>
>---

<br>



<br>

>---
>
>You can **append, or concatenate, DataFrames** with the `.concat()` function
>
>- The columns for the DataFrame should be identical
>
>- `pd.concat([df1, df2])` will stack the rows from `df2` at the bottom of `df1`
>
>---

<br>



```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_parquet('../Data/retail_2016_2017.parquet')
retail.head()
```

```python
automotive = retail.query("family == 'AUTOMOTIVE'")
automotive.head()
```

```python
beauty = retail.query("family == 'BEAUTY'")
beauty.head()
```

```python
# To append both new DataFrames we need to use "pd.concat()"
#
# NOTE: "ignore_index=False" will remain the original index of both DF
pd.concat([automotive, beauty], axis=0, ignore_index=False) 
```

```python
pd.concat([automotive, beauty], axis=0, ignore_index=True) # reset the index
```
