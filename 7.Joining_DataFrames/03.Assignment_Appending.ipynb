```python
import pandas as pd
import numpy as np
```

```python
transactions = pd.read_csv('../Data/transactions.csv',
                                      names=['Date', 'Store_Number', 'Transaction_Count'], 
                                      header=0,
                                      skiprows=0,
                                      parse_dates=['Date'],
                                      dtype={'Store_Number':'Int8',
                                             'Transaction_Count':'Int16'}
                            ).assign(target_pct = lambda x: x.Transaction_Count/2500,
                                     met_target = lambda x: x.target_pct >= 1,
                                     bonus_payable = lambda x: x.met_target * 100,
                                     month = lambda x: x.Date.dt.month,
                                     day_of_week = lambda x: x.Date.dt.day_of_week,
                                     ).astype({'target_pct':'Float32',
                                               'month':'Int8',
                                               'day_of_week':'Int8'})
transactions
```

```python
transactions_2014 = transactions.query("Date.dt.year == 2014")
display(transactions_2014.head())

transactions_2015 = transactions.query("Date.dt.year == 2015")
transactions_2015.head()
```

```python
df = pd.concat([transactions_2014, transactions_2015], 
               axis=0, 
               ignore_index=True)

display(df.head())
df.tail()
```
