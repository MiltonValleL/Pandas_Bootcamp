## Calculating the Mean using Different Methods

```python
original_mean = df.mean().iloc[0]
ffill_mean = df.ffill().mean().iloc[0]
bfill_mean = df.bfill().mean().iloc[0]
interpolation = df.interpolate().mean().iloc[0]

# Show results
print('-'*60)
print(f"  - Mean with Missing Values: {original_mean:.2f}\n")
print(f"  - Mean with Forward Fill: {ffill_mean:.2f}\n")
print(f"  - Mean with Backfill: {bfill_mean:.2f}\n")
print(f"  - Mean with Interpolation: {interpolation:.2f}")
print('-'*60)
```
