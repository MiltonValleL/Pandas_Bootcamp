```python

df = transactions.assign(
    year = transactions.date.dt.year,
    month = transactions.date.dt.month,
    day_of_week = transactions.date.dt.day_of_week,
    time_to_last_date = (transactions.date.max() - transactions.date) + pd.to_timedelta(3, unit='W'),
    weeks_to_last_date = lambda x: (x.time_to_last_date.dt.days)/7
)

df
```
