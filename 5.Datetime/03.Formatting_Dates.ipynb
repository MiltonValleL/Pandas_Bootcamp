# FORMATTING DATES

---

| **CODE** | **EXAMPLE** | **DESCRIPTION** | BELONG TO |
|:---|:---|:---|:---|
| `%D` | 07/14/2022 | Zero-padded date (based on OS settings | Date |
| `%Y` | 2022 | Four digit year | Date |
| `%m` | 07 | Month of the year | Date |
| `%B` | July | Full month name (*based on OS settings*) | Date |
| `%d` | 14 | Day of the month |  Date |
| `%` | 4 | Weekday number (0-6, Sunday-Saturday) |  Date |
| `%A` | Thursday | Full weekday name (*based on OS settings*) |  Date |
| `%a` | Thu | Short weekday name (*based on OS settings*) |  Date |
| `%U` | 28 | Week of the year (*starting on the first sunday*) |  Date |
| `%j` | 195 | Day of the Year |  Date |
| `%T` | 16:36:30 | Zero-padded time (*24-hour format*) |  Time |
| `%H` | 16 | Hour (*24-hour format*) |  Time |
| `%I` | 04 | Hour (*12-hour format*) |  Time |
| `%p` | PM | AM or PM (*based on OS settings*) |  Time |
| `%M` | 36 | Minute |  Time |
| `%S` | 30 | Second) |  Time |

>---
>
>The `dt.accessor` lets you use `strftime()` and `strptime` on datetimes in Pandas DataFrames
>
>- <span style='color:orange'>`strftime` stands for **String Format Time** (*converts a datetime object to a string*).
>
>- <span style='color:red'>`strptime` stands for **String Parse Time** (*converts a string to a datetime object*).

```python
# copy the column 'date' as a DataFrame
df = retail[['date']].copy()

# Create variation columns of 'date'
#`%D`: Zero-padded date
#`%Y`: Four digit - Year
#`%m`: Two-digit - Month of the year
#`%B`: Full month name
#`%b`: 3-letter month name
#`%d`: Two-digit - Day of the month |  Date |
#`%`:  Weekday number (0-6, Sunday-Saturday)
#`%A`: Full weekday name
#`%a`: Short weekday name
#`%U`: Week of the year
#`%j`: Day of the Year
```

```python
df = (df
    .assign(
        date_padded = df['date'].dt.strftime('%D'),  # (Object data type)

        
        # Standard .dt.strftime() doesn't recognize %q natively. 
        # To use it, you must first convert the datetime column to a period.
        year_quarter = df['date'].dt.to_period('Q').dt.strftime('%Y-Q%q'), # (Object data type)

        
        full_date = df['date'].dt.strftime('%a, %b-%d-%Y'), # (Object data type)
        FULL_DATE = df['date'].dt.strftime('%A, %B-%d-%Y'), # (Object data type)
        
        week_year = df['date'].dt.strftime('%U'), # (Object data type)
        WEEK_YEAR = df['date'].dt.strftime('Week: %U'), # (Object data type)
        
        day_year = df['date'].dt.strftime('%j'), # (Object data type)
        DAY_YEAR = df['date'].dt.strftime('Day: %j'), # (Object data type)

        week_day = df['date'].dt.strftime('%w'), # Sunday = 0, Saturday = 6  -  (Object data type)
        WEEK_DAY1 = df['date'].dt.day_of_week, # Monday = 0, Sunday = 6  -  # (int32 data type)
        WEEK_DAY2 = df['date'].dt.dayofweek, # Monday = 0, Sunday = 6  -  # (int32 data type)
    )
)

df
```
