```python
# find the MAX date registered
max_date = transactions['date'].max()
max_date
```

```python
transactions = (transactions
    .assign(
        time_to_last_date = max_date - transactions.date,
        year = transactions.date.dt.year,
        month = transactions.date.dt.month,
        day_of_week = transactions.date.dt.day_of_week)
)

transactions
```

```python
transactions['date'] = transactions['date'].dt.strftime('%Y-%B-%d')

transactions
```
