# TIME SERIES INDICES

---

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'date':['2022-01-01', '2022-01-02', '2022-02-01', '2022-03-01', '2023-01-01', '2023-01-02'],
                   'sales':[10, 15, pd.NA, 20, 30, 35]})
df['date'] = pd.to_datetime(df['date'], errors='coerce', format='%Y-%m-%d')

print(df.info())
print('-'*75)

df
```

```python
df[df['date'].dt.year == 2023]
```

<br>

>---
>Use **datetimes as the index** to allow for intuitive slicing of your DataFrame.
>
>- Like with integers, make sure your time index is sorted or you will get very odd results.

```python
# configure the 'date' column as index
df = df.set_index('date', verify_integrity=True)
df
```

```python
df.info()
```

## Filtering a DateTime index

```python
# When 'datetime' column is the index the filtering task is easier
df.loc['2023']
```

```python
# filtering year 2022
df.loc['2022']
```

```python
# filtering years 2022 and 2023
df.loc['2022':'2023']
```

```python
# filtering January and February from year 2022
df.loc['2022-01':'2022-02']
```

```python
# filtering more specific dates YYYY-MM-DD
df.loc['2022-01-01':'2023-01-01']
```
