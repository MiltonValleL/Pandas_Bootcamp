```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
oil = pd.read_csv('../Data/oil.csv', parse_dates=['date'], header=0, names=['date', 'price'])
oil.info()
```

```python
oil.head(3)
```

```python
df = oil.dropna()
df
```

```python
df = df.set_index('date')
df
```

```python
import seaborn as sns

sns.set_style('darkgrid')
```

```python
period = ['W', 'ME', 'QE', 'YE']
name = {'W':'Week', 'ME':'Month', 'QE':'Quarter', 'YE':'Year'}
color = ['green', 'orange', 'blue', 'brown']
line = ['--', '.-', '-.', ':']

for n in range(4):
    (df.resample(period[n])
        .mean()
        .plot(figsize=(10, 4),
              title=f'Assignment 5 - Period Analysis by "{name[period[n]]}"',
              xlabel=name[period[n]].upper(),
              grid=True,
              ylabel='Oil Mean Price [USD]',
              style=line[n],
              color=color[n]))
```
