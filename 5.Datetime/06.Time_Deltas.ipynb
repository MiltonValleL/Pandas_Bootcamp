# TIME DELTAS & ARITHMETIC

---

>---
>**Time Deltas** represent the amount or time, or difference, between two datetimes.
>
>- A time delta is returned when subtracting two datetime values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'order_date':['2021-04-04', '2021-05-23'],
                   'delivery_date':['2021-04-07', pd.NA]})

df = df.astype({'order_date':'datetime64[ns]', 'delivery_date':'datetime64[ns]'})
df.info()
```

```python
df['Ship_days'] = df.delivery_date - df.order_date # this operation create a `timedelta64[ns]`
df
```

```python
df['shipping_days'] = (df.delivery_date - df.order_date).dt.days # This operation create a `float64`
df
```

```python
df.info()
```

>---
>The `.to_timedelta()` method lets you use **frequencies** to return timedelta values
>

| **FREQUENCY** | **UNIT** |
|:---:|:---|
| `"W"` | Week |
| `"D"` | Day |
| `"h"` | Hour |
| `"min"` | Minute |
| `"s"` | Second |

```python
pd.to_timedelta(5, unit='W')
```

```python
pd.to_timedelta(5, unit='D')
```

```python
pd.to_timedelta(5, unit='h')
```

```python
pd.to_timedelta(5, unit='min')
```

```python
pd.to_timedelta(5, unit='s')
```

```python
pd.to_timedelta(5, unit='Y')
```

>---
>Because months, quarters, and years have differing lengths, they are sometimes known as "anchored" frequencies, since they are anchored (*anclado*) to the end of calendar periods.
>
>They can't be used in time delta arithmetic, but are useful for things like resampling.
>

| **FREQUENCY** | **UNIT** |
|:---:|:---|
|`"M"` | Month end |
|`"Q"` | Quarter end |
|`"Y"` | Year end |

```python
df
```

```python
# Adding 1 day to "order_date"
df.order_date + pd.to_timedelta(1, unit="D")
```

```python
# Adding 2 weeks to "order_date"
df.order_date + pd.to_timedelta(2, unit="W")
```

```python
retail = retail.assign(
    birthday=pd.to_datetime('1950-01-01') # This columns is created as datetime64[ns]
)

retail.head(3)
```

```python
retail = retail.assign(
    age_in_days = retail.date - retail.birthday,
    age_in_days_num = (retail.date - retail.birthday).dt.days,
    age_in_years = (retail.date - retail.birthday).dt.days / 365.25,
    retirement_date = (retail.date + pd.to_timedelta(365.25, unit='D') * 20)
)

retail
```
