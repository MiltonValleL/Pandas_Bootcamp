# AGGREGATION & RESAMPLING

---

<br>

>---
>You can **aggregate time series** using the `.groupby()` Method, but depending on use case there are better methods.

<br>

```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_parquet('../Data/retail_2016_2017.parquet')
retail.info()
```

```python
retail.head(3)
```

```python
df = retail[['date', 'sales']]
df
```

<br><br>

---

<br><br>

```python
# This code sums all the sales per month, including all the years of the dataset
df.groupby(df.date.dt.month).agg({'sales':'sum'}).round()
```

```python
# to filter per year we need to add the following...
df.groupby([df.date.dt.year, df.date.dt.month]).agg({'sales':'sum'}).round()
```

<br><br>

>---
>**Resampling** is a special form of time series aggregation that modifies date indices and fills any gaps to create continuous date values.
>
>- Frequencie are used to define the granularityu (*you can use years, quarters, and months now.*)

```python
df2 = df.set_index('date')

df2.resample('ME').sum().round()
```
