# CONVERTING TO DATETIMES

---

<br><br>

> The `.astype()` method can convert strings to datetimes, but has some **limitations**:
>
>- Any values that Pandas can't correctly identify as dates will return an error.

```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.DataFrame({'order_date':['2021-04-04', '2021-05-23'],
                   'delivery_date':['2021-04-07', pd.NA]})

print(df.info())
print('-'*60)

df
```

```python
df = df.astype({'order_date':'datetime64[ns]', 
           'delivery_date':'datetime64[ns]'})

print(df.info())
print('-'*60)

df
```

>---
>`.astype()` METHOD
>
>The `.astype()` method can convert strings to datetimes, but has som limitations:
>
>- Any values that Pandas can't correctly identify as dates will return an error
>
>- Ignoring the errors will return the Series as an `object` data type, not `datetime64[ns]`

<br><br>

>---
>`.to_datetime()` FUNCTION
>
>The `.to_datetime()` function is a more flexible option for converting datetimes

<br><br>

>---
>NOTE: `NaT` stands for **Not a Time**

```python
# The DataFrame
df = pd.DataFrame({'order_date':['2021-04-04', '2021-05-23'],
                   'delivery_date':['2021-04-07', pd.NA]})

# The DateTime Conversion
df['delivery_date'] = pd.to_datetime(df['delivery_date'], 
                                     errors='coerce',
                                     format='%Y-%m-%d',
                                    )
df
```
