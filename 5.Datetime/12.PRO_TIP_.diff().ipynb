# PRO TIP: `.diff()`

---

>---
>The `.diff()` Method calculates the difference between the values in a Series and those same values shifted a specified number of periods
>
>- This is useful in measuring absolute changes over time
>  
>---

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'sales':[10, 15, 15, 20, 30, 35]},
                 index=pd.date_range("2022-01-01", "2022-01-06"))

df
```

```python
df.diff()
```

```python
# .diff() Method is equivalent to...
df - df.shift(1)
```

```python
df
```

```python
df.diff(2)
```

## LABORATORY

```python
retail = pd.read_parquet('../Data/retail_2016_2017.parquet')
retail
```

```python
df = retail.groupby('date').agg({'sales':'sum'}).round().iloc[1:10]
df
```

```python
df.assign(
    daily_change = df.sales.diff(),
    growth_percentage = ((df.sales.diff() / df.sales.shift())*100).round(2)
)
```
