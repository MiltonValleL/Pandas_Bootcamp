# DATE AND TIME PARTS

---

<br><br>

>---
>While you can also use the `.strftime()` method to **extract datetime components**, it's prefered to use the dedicated `.dt` **accessors**.

```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_parquet('../Data/retail_2016_2017.parquet')
retail.head()
```

```python
retail.info()
```

```python
# Show the date
retail['date'].dt.date
```

```python
# Show the year
retail['date'].dt.year
```

```python
# Show the month
retail['date'].dt.month
```

```python
# Show de day of the week
retail['date'].dt.dayofweek
```

```python
# Show time
retail['date'].dt.time
```

```python
# Show the Hour
retail['date'].dt.hour
```

```python
# Show the Minute
retail['date'].dt.minute
```

```python
# Show the Seconds
retail['date'].dt.second
```

```python
df = retail[['date']]
df
```

```python
df = df.assign(
    year = df['date'].dt.year,
    quarter = df['date'].dt.quarter,
    day_of_week = df['date'].dt.dayofweek,
)

df
```
