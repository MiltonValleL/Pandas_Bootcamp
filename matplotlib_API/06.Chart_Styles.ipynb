```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv', parse_dates=['date'])

retail = retail[(retail
    .family.isin(list(retail.family.unique()[8:12])))
    & (retail.store_nbr == 18) 
    & (retail.date.dt.year == 2017) 
    & (retail.date.dt.month == 4)][['date', 'family', 'sales']]

retail
```

```python
import seaborn as sns

sns.set_style('darkgrid')

(retail
    .pivot_table(index='date', columns='family', values='sales', aggfunc='sum')
    .plot(title='Total Sales of Store NO. 18 on April 2017',
          xlabel='DATE', 
          ylabel='Total Sales [USD]',
          grid=True,
          style=[':', '-.', '.-', '-']).legend(bbox_to_anchor=(1.4, 1))
)
```

## LABORATORY

```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date']).set_index('date')

stores = transactions.query("store_nbr in [15, 30] and date.dt.year == 2017").pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='sum')

# ---------------------------------------------------------------
sns.set_style('darkgrid')

stores.plot(title='Total Sales for Stores 15 and 30',
            xlabel='DATE',
            ylabel='Total Number of Transactions',
            color=['blue', 'orange'],
            alpha=0.9,
            grid=True,
            style=['-.', '.-']).legend(bbox_to_anchor=(1.2,1), loc='center right');
```
