# HISTOGRAMS

---

>**Histograms** are used for showing the distribution of numerical series.
>
>
>- It divides the data in `bins` and plots the frequency of values that fall into each bin as bars

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv', parse_dates=['date'])

sales_df = (retail
            .query("sales > 0 and date.dt.year == 2016 and date.dt.month == 3")
            .pivot_table(index='date', columns='family', values='sales', aggfunc='sum').round()
           )
sales_df.head()
```

```python
sales_df['PREPARED FOODS'].plot.hist(
    bins=10,
    title='Prepared Foods Histogram',
    xlabel='Total Sales',
)
```

```python
sales_df[['PREPARED FOODS', 'FROZEN FOODS']].plot.hist(
    bins=12,
    title='Prepared Foods & Frozen Foods - Histogram',
    xlabel='Total Sales',
    alpha=0.4 # helpful with more than one variable
);
```
