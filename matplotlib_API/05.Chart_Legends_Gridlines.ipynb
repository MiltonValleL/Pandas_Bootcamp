```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv', parse_dates=['date'])
retail = retail[(retail.family.isin(list(retail.family.unique()[8:12]))) & (retail.store_nbr == 18) & (retail.date.dt.year == 2017) & (retail.date.dt.month == 4)]
retail
```

```python
df_plot = retail.pivot_table(index='date', columns='family', values='sales', aggfunc='mean')
df_plot.head(3)
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             style=['-.', ':', '-', '--'],
             grid=True);
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             style=['-.', ':', '-', '--'],
             legend=False);
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             grid=True,
             style=['-.', ':', '-', '--']).legend(loc='best');
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             style=['-.', ':', '-', '--']).legend(loc='upper left');
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             grid=True,
             style=['-.', ':', '-', '--']).legend(loc='upper center');
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             style=['-.', ':', '-', '--']).legend(loc='lower right');
```

```python
df_plot.plot(title='Total Sales on April by Store NO. 18',
             xlabel='DATE',
             ylabel='Total Sales [USD]',
             color=['red', 'blue', 'green', 'orange'],
             alpha=0.9,
             grid=True,
             style=['-.', ':', '-', '--']).legend(bbox_to_anchor=(1.4, 1));
```

```python
transactions = pd.read_csv('Data/transactions.csv', parse_dates=['date']).set_index('date')

stores = transactions.query("store_nbr in [15, 30] and date.dt.year == 2017").pivot_table(index='date', columns='store_nbr', values='transactions', aggfunc='sum')
stores.plot(title='Total Sales for Stores 15 and 30',
            xlabel='DATE',
            ylabel='Total Number of Transactions',
            color=['blue', 'orange'],
            alpha=0.9,
            grid=True,
            style=['-.', '.-']).legend(loc='lower right');
```
