# DUPLICATED INDEX VALUES

---

>It is possible to have **duplicate index values** in a Pandas Series or DataFrame
>  - Accessing this indices by their label using `.ilo[]` returns all corresponding rows
>
>---
>You can **reset the index** in a Pandas Series or DataFrame back to the default range of integers by using the `.reset_index()` method
>   - By default, the existing index will become a new column in a DataFrame

```python
import numpy as np
import pandas as pd
```

```python
sales = [0, 5, 155, 0, 518]
items = ["coffee", "coffee", "tea", "coconut", "sugar"]

sales_series = pd.Series(data=sales, index=items, name="Sales")
sales_series
```

```python
sales_series.loc["coffee"]
```

```python
sales_series = sales_series.reset_index()
sales_series
```

```python
oil = pd.read_csv("Data/oil.csv").dropna()
oil.head()
```

```python
dates = np.array(oil["date"].iloc[1000:1100])
oil_array = np.array(oil["dcoilwtico"].iloc[1000:1100])

oil_series = pd.Series(data=oil_array, index=dates, name="oil_price")
oil_series
```

```python
# Calculate the mean of the first 10 and last 10 values
mean_first = oil_series.iloc[:10].mean()
mean_last = oil_series.iloc[-10:].mean()

print(f"mean firts 10 rows: {mean_first}\nmean last 10 rows: {mean_last}")
```

```python
# Grab oil prices from January 1st, 2017 to January 7th, 2017.
oil_slice = oil_series.loc["2017-01-01":"2017-01-07"]
oil_slice
```

```python
# Then revert the index of this slice back to integers
oil_slice = oil_slice.reset_index(drop=True)
```
