### Perform Mathematical Operations with Pandas Methods

```python
my_series = pd.Series(
    data=[1, np.nan, 2, 3, 4], index=["day 0", "day 1", "day 2", "day 3", "day 4"]
)
my_series
```

```python
# Add a number
my_series.add(2)
```

```python
# --------------------------------------
# Add a number and fill missing values
# --------------------------------------
my_series2 = my_series.add(0, fill_value=0).astype("int")
my_series2
```

```python
# Multiply by a number
my_series2.mul(3)
```

```python
# subtract from a number
my_series2.subtract(1)
```

```python
# Divide by number
my_series2.div(2)
```

```python
# Peform a Floor division
my_series2.floordiv(2)
```

```python
# Perform a modulo division
my_series2.mod(2)
```

```python
# Perform a Power operation
my_series2.pow(3)
```

---

### We can perform the same operation between Series

```python
my_series.add(my_series2)
```

```python
my_series2.mul(my_series)
```

```python
my_series.pow(my_series2)
```
