# THE SERIES INDEX & CUSTOM INDICES

---

```python
import numpy as np
import pandas as pd
```

```python
sales = [0, 5, 155, 0, 518]
items = ["coffee", "bananas", "tea", "coconut", "sugar"]

# Indices on creation of a Series
sales_series = pd.Series(data=sales, index=items, name="Sales")
sales_series
```

```python
# Indices by assign
sales_series.index = ["air", "candy", "snickers", "sun", "jeans"]
sales_series
```

```python
# Access values from index
sales_series["jeans"]
```

```python
# Slicing with name index - Last value is INCLUSIVE
sales_series["candy":"jeans"]
```

```python
# Slicing with Value index - Last value is NOT INCLUSIVE
sales_series[1:4]
```
