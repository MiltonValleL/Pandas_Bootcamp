# ARRAY FUNCTIONS

---

```python
import numpy as np
```

```python
sales = np.array([[0, 5, 155, 0, 518], [0, 1827, 616, 317, 325]])
sales
```

```python
np.median(sales)
```

```python
np.mean(sales)
```

### Percentiles

```python
np.percentile(sales, 50)
```

```python
np.percentile(sales, 90)
```

### Unique Values

```python
np.unique(sales)
```

### Square Root

```python
np.sqrt(sales)
```

## LABORATORY: Array Functions

```python
price = np.array([2.6, 7.2, 6.92, 3.88, 0.78, 2.67, 8.87, 0.23, 5.23, 6.3])

inventory = np.array([79, 22, 20, 48, 8, 76, 12, 16, 5, 64])

product_value = price * inventory
product_value
```

```python
product_value.sort()
product_value
```

```python
product_value
```

```python
reshaped_product_value = product_value.reshape(2, 5)
reshaped_product_value
```

---

```python
np.median(reshaped_product_value)
```

```python
np.percentile(reshaped_product_value, 50)
```

```python
np.percentile(reshaped_product_value, 90)
```

```python
np.percentile(reshaped_product_value, 0)
```

```python
np.percentile(reshaped_product_value, 100)
```

---

```python
np.sqrt(reshaped_product_value)
```

```python
np.unique(reshaped_product_value)
```

`NOTE`: **Unique Function** tends to be more useful when working with categorical data
