# FILTERING ARRAYS & MODIFYING ARRAY VALUES

---

## Filtering Arrays

```python
import numpy as np
```

```python
sales_array = np.array([[0, 5, 155, 0, 518], [0, 1827, 616, 317, 325]])
sales_array
```

```python
# Evaluate values different from 0
sales_array != 0
```

```python
# Show values different from 0
sales_array[sales_array != 0]
```

---

### Multiple Logical Tests

```python
# Multiple Logical Tests - OR
sales_array[(sales_array == 616) | (sales_array < 100)]
```

```python
# Multiple Logical Tests - AND
sales_array[(sales_array > 100) & (sales_array < 500)]
```

```python
# PRO TIP: Store complex filtering criteria in a variable

# Working with Boolean Masks
mask = (sales_array > 100) & (sales_array < 500)

sales_array[mask]
```

---

### Filtering arrays based on values of other arrays

```python
# array 1
sales_array[0, :]
```

```python
# Array 2
product_array = np.array(["fruits", "vegetables", "cereal", "dairy", "eggs"])
product_array
```

```python
# Filtering 2 using 1
product_array[sales_array[0, :] > 0]
```

---

### Modify Array Values

```python
a = sales_array[0, :]
a
```

```python
# Method 1: Used to modify a single parameter
a[1] = 30
a
```

```python
# Method 2: Used to modify many many parameters with the same value
a[a == 0] = -6
a
```

## LABORATORY : Filtering and Modifying Arrays

```python
my_array = np.arange(20)
my_array
```

```python
# Show only even numbers
my_array[my_array % 2 == 0]
```

```python
even_odd = np.array(["even", "odd"] * 10)
even_odd
```

```python
# Show only 'odd' values
even_odd[even_odd == "odd"]
```

```python
# Change even numbers for 0
print(my_array)

my_array[even_odd == "even"] = 0
my_array
```

```python
my_array[2] = 2
my_array
```
