# PARSING DATES & DATA TYPES

---

>---
>Dates are read in as object data types by default, buy you can **parse dates** with the `parse_dates` argument to convert them to `datetime64[ns]`
>
>- Specifying `infer_datetime_format=True` will speed up the date parsing

```python
import pandas as pd
import numpy as np
```

```python
# First look of the dataset
oil = pd.read_csv('Data/oil_expanded.csv')
oil.head(15)
```

```python
# Now it's time to parse dates
df = pd.read_csv('Data/oil_expanded.csv',
                 parse_dates=['Date'],
                 na_values=['MISSING', '.', '..'],
                )

df.head(15)
```

```python
df.info()
```

---

>---
>You can set the **data type** for each column with the `dtype` argument by passing in a dictionary with column names as keys and desired data types as values.
>
>- Get your data into its most efficient format from the start

```python
# Now it's time to parse dates
df = pd.read_csv('Data/oil_expanded.csv',
                 parse_dates=['Date'],
                 na_values=['MISSING', '.', '..'],
                 dtype={'Oil_Price':'float32',
                        'Sales':'int16'
                       }
                )

display(df.head(15))

df.info()
```
