```python
transactions = pd.read_csv('Data/transactions.csv', 
                          names=['Date', 'Store_Number', 'Transaction_Count'],
                          header=0,
                          parse_dates=['Date'],)
display(transactions.head())

transactions.info(memory_usage='deep')
```

```python
# We will obtain the same results as the upper cell. But from the ".read_csv()" Function

transactions_assignment = pd.read_csv('Data/transactions.csv',
                                      names=['Date', 'Store_Number', 'Transaction_Count'], 
                                      header=0,
                                      skiprows=0,
                                      parse_dates=['Date'],
                                      dtype={'Store_Number':'Int8',
                                             'Transaction_Count':'Int16'}
                            ).assign(target_pct = lambda x: x.Transaction_Count/2500,
                                     met_target = lambda x: x.target_pct >= 1,
                                     bonus_payable = lambda x: x.met_target * 100,
                                     month = transactions.Date.dt.month,
                                     day_of_week = transactions.Date.dt.day_of_week,
                                     ).astype({'target_pct':'Float32',
                                               'month':'Int8',
                                               'day_of_week':'Int8'})
display(transactions_assignment)

transactions_assignment.info(memory_usage='deep')
```
