# CONVERTERS - PRO TIP

---

>---
>You can apply functions to columns of data by using the `converters` argument.
>
>- Pass a dictionary with the column names as keys and the functions as values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/oil.csv',
                names=['Date', 'Oil_Price'],
                header=0)
df.head(15)
```

```python
# create a column "sales" that increases it's value by 1 each cell
add_col = list(range(1, len(df)+1))

df['Sales'] = add_col
df
```

```python
df.to_csv('Data/oil_expanded.csv', index=False)
```

```python
df = pd.read_csv('Data/oil_expanded.csv')
df.head(15)
```

```python
# Fills wrong values with Zero
missing = lambda x: x if x not in ['MISSING', '.', '..'] else 0

# Adds "$" char and 2 decimal zeros
currency = lambda x: f"${x}.00"

# Now we import again our DataFrame and fix the issues
df = pd.read_csv('Data/oil_expanded.csv', 
                 converters={'Oil_Price': missing,
                             'Sales': currency}
                )

df.head(15)
```
