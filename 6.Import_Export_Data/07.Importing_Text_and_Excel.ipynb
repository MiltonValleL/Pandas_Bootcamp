## Reading an EXCEL File

>---
>The `.read_excel()` function is used to **read in Excel files** in Pandas
>
>- You can specify worksheets by passing the sheet name or position to the `sheet_name` argument

```python
# Leer el primer sheet (Sheet 1)
df1 = pd.read_excel('Data/monthly_sales.xlsx', sheet_name=0)

# Leer el segundo sheet (Sheet 2) - 0-indexed
df2 = pd.read_excel('Data/monthly_sales.xlsx', sheet_name=1)

display(df1)

df2
```

```python
# O acceder por nombre
df1 = pd.read_excel('Data/monthly_sales.xlsx', sheet_name='Sheet 1')
df2 = pd.read_excel('Data/monthly_sales.xlsx', sheet_name='Sheet 2')

display(df1)

df2
```

```python
# Leer TODOS los sheets como diccionario
all_sheets = pd.read_excel('Data/monthly_sales.xlsx', sheet_name=None)
# all_sheets = {'Sheet 1': DataFrame, 'Sheet 2': DataFrame}

all_sheets
```

<br><br>

>---
>You can use Pandas' `.concat()` function to **append data from multiple sheets


```python
all_sheets = pd.concat(
    pd.read_excel('Data/monthly_sales.xlsx', sheet_name=None), 
    ignore_index=True
)

all_sheets
```


>- `sheet_name=None` makes pandas read every sheet and store them as a dictionary, which is appended using `.concat()`.
>
>- `ignore_index=True` makes sure each row has a unique index.
