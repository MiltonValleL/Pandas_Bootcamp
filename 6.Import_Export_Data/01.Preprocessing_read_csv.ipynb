## `.read_csv()` Revisited

>---
>
>The ***.read_csv()*** function only needs a file path to read in a file, but it also has many capabilities fro preprocessing the data using other arguments:
>

| **PARAMETER** | **=** | **SETTTING** | **DESCRIPTION** |
|---:|:---:|:---|:---|
| `file` | = | <span style='color:orange'>***path/name.csv***</span> | File path & name to read in (*can also be a URL*) |
| `sep` | = | <span style='color:orange'>***/***</span> | Character used to separate each column of values (*default is comma*) |
| `header` | = | <span style='color:orange'>***0*** | Row number to use as the column names (*default is "infer"*) |
| `names` | = | <span style='color:orange'>***["date", "sales"]*** | List of column names to override the existing ones |
| `index_col` | = | <span style='color:orange'>***"date"*** | Column name to be used as the DataFrame index |
| `usecols` | = | <span style='color:orange'>***["date", "sales"]***</span> | List of columns to keep in the DataFrame |
| `dtype` | = | <span style='color:orange'>***{"date":"datetime64[ns]", "sales":"Int32"}*** | Dictionary with column names and data types |
| `parse_dates` | = | <span style='color:orange'>***True*** | Converts date strings into datetimes when True |
| `infer_datetime_format` | = | <span style='color:orange'>***True*** | Makes parsing dates more efficient |
| `na_values` | = | <span style='color:orange'>***["-", "#N/A"]*** | Strings to recognize as **NaN values** |
| `nrows` | = | <span style='color:orange'>***100*** | Number of rows to read in |
| `skiprows` | = | <span style='color:orange'>***[0,2]*** | Line numbers to skip *(accepts lambda functions)* |
| `converters` | = | <span style='color:orange'>***{"sales":lambda x: f"${x}"}*** | Dictionary with column names and functions to apply |

---

>---
>Pandas will try to infer the **Column Names** in your file by default, but there are several options to override this behavior:

```python
import pandas as pd
import numpy as np
```

```python
# - The first row is used for the column headers by default.
# - A suffix of ".1" is added to avoid duplicated column names

df = pd.read_csv('../Data/monthly_sales.csv')
df
```

```python
# Specify "header=None" to keep all the rows in the file as data, and use integers as column names
pd.read_csv('../Data/monthly_sales.csv', header=None)
```

```python
# Specify "header=0" and use the names argument to pass a list of desired column names
pd.read_csv('../Data/monthly_sales.csv', header=0)
```

```python
# We can set the column names before importing the CSV file...
cols = ['Date', 'grocety_sales', 'beer_sales', 'to_drop']

pd.read_csv('../Data/monthly_sales.csv', header=0, names=cols)
```
