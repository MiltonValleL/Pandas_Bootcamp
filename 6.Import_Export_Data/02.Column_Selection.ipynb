# COLUMN SELECTION - SETTING AN INDEX

---

>---
>You can **set the index** for the DataFrame with the `index_col` argument
>
>- Pass a list of column names to create a multi-index DataFrame
>
>- Specify `parse_dates=True` to convert index date column to a datetime data type

```python
import pandas as pd
import numpy as np
```

<br><br>

---

## 1. Setting the Index

```python
# Setting "index_col='date'" will set date as a STRING
df = pd.read_csv('Data/oil.csv', index_col='date')
df
```

```python
type(df.index)
```

```python
# If we want to filter by Year. We get an error
df.loc['2014']
```

```python
# Now if we set "parse_date=True" and "index_col='date'" we do get an datetime64[ns] index
df = pd.read_csv('Data/oil.csv', index_col='date', parse_dates=True)
df
```

```python
type(df.index)
```

```python
# Now if we want to filter by year again
df.loc['2014']
```

## 2. Selecting Columns

<br>

>---
>You can **select the columns** to read in with the "use_cols" argument
>
>- This can save a lot of processing time and memory


```python
# The original DataFrame has the following columns...
df = pd.read_csv('Data/retail_2016_2017.csv')
df
```

```python
df.info(memory_usage='deep')
```

```python
# Now we'll work only with the columns "date" and "sales" the rest of the columns is noise for this case...
df = pd.read_csv('Data/retail_2016_2017.csv', usecols=['date', 'sales'])
df.head()
```

```python
df.info(memory_usage='deep')
```

> If we contrasts the **Memory Usage** from the first and last import. We'll see that the **Memory Usage** has reduced more than the 50%.
>
> Which besides the noise of the useless columns is a great practice to do it while you import the CSV file.
