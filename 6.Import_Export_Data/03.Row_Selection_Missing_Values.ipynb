# ROW SELECTION & MISSING VALUES

---

<br><br>

>---
>You can **select the rows** to read in from the top with the `nrows` argument, and specify any rows to skip with `skiprows`

```python
import pandas as pd
import numpy as np
```

```python
# original Dataset...
df = pd.read_csv('Data/transactions.csv')
df
```

---

## Working with `nrows=`

```python
# Fixed Dataset using "nrows"
# "nrows" is used for reading just pieces of large files
df = pd.read_csv('Data/transactions.csv', nrows=5)
df
```

```python
len(df)
```

---

## Working with `skiprows=`

```python
df = pd.read_csv('Data/transactions.csv', skiprows=[1, 3, 5])
df
```

```python
# if we set "skiprows=0" will remove the column names, so we'll need to configure it. 
# Or we'll have the first data-row as the "Column Name"...
# **REMEMBER THIS**
df = pd.read_csv('Data/transactions.csv', skiprows=[0, 2, 4])
df
```

```python
# skip all the odd rows
df = pd.read_csv('Data/transactions.csv', skiprows=lambda x: x%2 == 1)
df
```

## Missing Values

<br>

>---
>You can specify strings (*or other values*) to treat as **missing values** with the `na_values` argument.
>
>- They are replaced with Numpy `NaN` values

```python
# Original Dataset
df = pd.read_csv('Data/oil.csv')
df.head(15)
```

```python
# Fixed Dataset with wrong values, converted into Missing Values
df = pd.read_csv('Data/oil.csv', na_values=['MISSING', '.', '..']) # "na_values" accepts also numbers
df.head(15)
```
