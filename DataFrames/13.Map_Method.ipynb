# THE MAP METHOD

---

The `.map()` method **maps the values to a column or an entire DataFrame**

    - You can pass a dictionary with existing values as the keys, and new values as the values

```python
import numpy as np
import pandas as pd
```

```python
product_df = pd.DataFrame({'product_id':[1, 2, 3, 4, 5],
                           'product' : ['Dairy', 'Dairy', 'Dairy', 'Vegetables', 'Fruits'],
                           'price': [2.56, 2.56, 4.55, 2.74, 5.44]
                          })
product_df
```

```python
# EXAMPLE NO.1 - Mapping in a new column

mapping_dict = {'Dairy': 'Non-Vegan', 'Vegetables':'Vegan', 'Fruits':'Vegan'}

product_df1 = product_df.copy()
product_df1['Vegan?'] = product_df1['product'].map(mapping_dict)
product_df1
```

```python
# EXAMPLE NO.2 - Mapping on the same column
product_df2 = product_df.copy()

product_df2['price'] = product_df2['price'].map(lambda x: f'${x}')
product_df2
```

```python
retail = pd.read_csv('Data/retail_2016_2017.csv')
retail.head()
```

```python
retail.family.value_counts()
```

```python
product_category_dict = {
    'PRODUCE': 'GROCERY',
    'POULTRY': 'GROCERY',
    'GROCERY I': 'GROCERY',
    'GROCERY II': 'GROCERY',
    'EGGS': 'GROCERY'
}

retail.family.map(product_category_dict).value_counts(dropna=False)
```

## How to solve this mapping problem - possible solution

---

`PRO TIP`: Be sure to MAP all the categories of `family` column. Otherwise you'll include several `NaN` values to all categories not specified on the original column

---

```python
mv_dict = {}

for value in retail.family.values:
    if value in product_category_dict.keys():
        mv_dict[value] = product_category_dict[value]
    else:
        mv_dict[value] = value

# now we'll map again...
retail.family.map(mv_dict).value_counts()
```
