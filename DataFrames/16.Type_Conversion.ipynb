# TYPE CONVERSION

---

Like a Series, you can convert data types in a DataFrame by using the `.astype()` method, and specifying the desired data type (*if compatible*)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Data/retail_2016_2017.csv')
sample_df = df.sample(5, random_state=2023)
sample_df
```

---

## Data Type Coversion on One Column

```python
sample_df['sales_int'] = sample_df['sales'].astype('int')
sample_df
```

```python
sample_df.info()
```

---

## Data Type Conversion on Multiple Columns

```python
df = pd.read_csv('Data/retail_2016_2017.csv')
sample_df = df.sample(5, random_state=2026)


sample_df = sample_df.astype({'date':'datetime64[ns]', 'onpromotion':'float', 'family':'category'})
sample_df
```

```python
sample_df.info()
```

---

```python
product_df = pd.DataFrame({'product_id':[1, 2, 3, 4, 5],
                           'product' : ['Dairy', 'Dairy', 'Dairy', 'Vegetables', 'Fruits'],
                           'price': ['$2.56', '$2.56', '$4.55', '$2.74', '$5.44']
                          })
product_df
```

```python
# converting an string data type into a column, after some cleaning
product_df.assign(price = product_df.price.str.strip('$').astype('float'))
```
