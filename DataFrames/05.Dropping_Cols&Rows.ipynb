---

You'll tipically drop rows via `slicing` or `filtering`, but it's worth being aware that `.drop()` method can be used as well.

---

```python
oil = pd.read_csv('Data/oil.csv')
oil.columns = ['date', 'price']
oil['euro_price'] = oil['price'] * 1.1
oil
```

---

`PRO TIP:` It's a good practice to create a new DataFrame when dropping columns from the original DF

---

```python
# Create a new DF to drop columns
oil_euro = oil.drop(labels='price', axis=1)
oil_euro
```

```python
# Original DF is complete
oil.head()
```

```python
# Now you can work on the new DF. And if you make errors - you can always come back to the original DF
oil_euro.drop(labels=[0, 2, 4, 6, 7, 9, 12], axis=0, inplace=True)
oil_euro
```

```python
# The reset the index in this case
oil_euro.reset_index(drop=True, inplace=True)
oil_euro
```
