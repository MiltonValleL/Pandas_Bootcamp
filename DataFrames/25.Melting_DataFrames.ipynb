# MELTING DATAFRAMES

---

The `.melt()` method will ***unpivot a DataFrame***, or convert columns into rows. This is great for turning a "wide" data table into a "long" format.

```python
import numpy as np
import pandas as pd
```

```python
country_revenue = pd.DataFrame({
    'country':['Algeria', 'Egypt', 'Nigeria', 'South Africa'],
    '2000':[151, 204, 277, 190],
    '2001':[171, 214, 301, 211],
    '2002':[176, 227, 311, 224],
    '2003':[184, 241, 342, 243]
})
country_revenue
```

---

## 1. Simple use of `.melt()` Method

```python
country_revenue.melt()
```

```python
country_revenue
```

---

## 2. Using Parameters of `.melt()` Method

```python
country_revenue.melt(id_vars='country')
```

---

We can also select the columns to melt and name the `variable` & `value` columns

```python
country_revenue
```

```python
country_revenue.melt(id_vars='country',
                     value_vars=['2001', '2002', '2003'],
                     var_name='year',
                     value_name='GDP',)
```

## LABORATORY

```python
pm = pd.read_excel('Data/premier_league_games_full.xlsx')
pm.head()
```

```python
pm = (pm.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']")
    .pivot_table(index='HomeTeam', columns='season', values='HomeGoals', aggfunc='mean')
    )
pm
```

```python
# we need to first reset the index. Otherwise we loose "HomeTeam"...
pm.melt()
```

```python
# If we "reset index" we have the HomeTeam
pm.reset_index().melt()
```

```python
# And the melting will successfully show...
pm.reset_index().melt(id_vars='HomeTeam')
```

```python
# Now we add 'value_vars'
pm.reset_index().melt(id_vars='HomeTeam',
                      value_vars=['2008/2009', '2009/2010', '2010/2011'],
                     )
```

```python
pm.reset_index().melt(id_vars='HomeTeam',
                      value_vars=['2008/2009', '2009/2010', '2010/2011'],
                      value_name='avg_goals')
```

## ASSIGNMENT: PIVOT & MELT



```python
transactions = pd.read_csv('Data/transactions.csv')

transactions = transactions.assign(
    target_pct = transactions.transactions / 2500,
    met_target = (transactions.transactions / 2500) >=1,
    bonus_payable = ((transactions.transactions / 2500) >=1) * 100,
    month = transactions.date.astype('datetime64[ns]').dt.month,
    day_of_week = transactions.date.astype('datetime64[ns]').dt.dayofweek)
    
transactions.head(3)
```

```python
transactions = transactions.assign(
    store_avg_trans = transactions.groupby(['store_nbr', 'day_of_week'])['transactions'].transform('mean'),
    trans_vs_avg = lambda x: x['transactions'] - x['store_avg_trans']
)
transactions
```

---
---

```python
pivot_transactions = (transactions
    .query("store_nbr <= 11 and bonus_payable > 0")
    .pivot_table(index='store_nbr', columns='day_of_week', 
                 values='bonus_payable', aggfunc='sum')
    .style.format("{:.2f}")
    .background_gradient(cmap='RdYlGn', axis=1)
)

pivot_transactions
```

```python
melt_transactions = (transactions
    .query("store_nbr <= 11 and bonus_payable > 0")
    .pivot_table(index='store_nbr', columns='day_of_week', 
                 values='bonus_payable', aggfunc='sum')
    .reset_index()
    .melt(id_vars='store_nbr',value_name='bonus_payable')
)

melt_transactions
```
