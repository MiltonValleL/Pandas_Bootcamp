```python
import pandas as pd
import numpy as np
```

---

## Renaming Columns

```python
product_df = pd.DataFrame({'product':['Dairy', 'Dairy', 'Dairy', 'Vegetables', 'Fruits'],
                           'price':[2.56, 2.56, 4.55, 2.74, 5.44]})
product_df
```

```python
# Now make it uppercase - List Comprehension
product_df.columns = [col.upper() for col in product_df.columns]
product_df
```

```python
product_df = pd.DataFrame({'product':['Dairy', 'Dairy', 'Dairy', 'Vegetables', 'Fruits'],
                           'price':[2.56, 2.56, 4.55, 2.74, 5.44]})

# Changing the names of the columns - METHOD 1
product_df.columns = ['product_name', 'cost']
product_df
```

```python
# Now capitalize - Lambda function
product_df.rename(columns=lambda x: x.upper())
```

```python
product_df = pd.DataFrame({'product':['Dairy', 'Dairy', 'Dairy', 'Vegetables', 'Fruits'],
                           'price':[2.56, 2.56, 4.55, 2.74, 5.44]})

# Changing the names of the columns - METHOD 2
product_df.rename(columns={'product':'product_name2', 'price':'cost2'})
```

## Reordering Columns

```python
product_df['product_id'] = range(1, 6)
product_df
```

```python
product_df.reindex(labels=['product_id', 'product', 'price'], axis=1)
```

```python
# For permanent change we need to asign otherwise work definetely
product_df
```

```python
product_df = product_df.reindex(labels=['product_id', 'product', 'price'], axis=1)
product_df
```
