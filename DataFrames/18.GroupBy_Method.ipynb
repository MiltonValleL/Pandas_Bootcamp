# 1. Working with `.groupby()` Method



>**Grouping a DataFrame** allows you to aggregate the data at a different level.
>
>   - For example, transform daily data into monthly, roll up transactions level data by store, etc...

>To group data, use the `.groupby()` method and specify a column to group by
>
>   - The grouped column becomes the index by default

````python
retail = pd.read_csv('Data/retail_2016_2017.csv')
retail.head()
````

````python
print('----------- This is a Series result ------------')
retail.groupby('family')['sales'].sum().head()
````

````python
print('----------- This is a DataFrame result ------------')
retail.groupby('family')[['sales']].sum().head()
````

# 2. Grouping By Multiple Columns

>You can **group by multiple columns** by passing the list of columns into `.groupby()`
>
>   - This creates a multi-index object with an index for each column the data was grouped by

````python
retail.head()
````

````python
retail.groupby(['family', 'store_nbr'])[['sales']].sum()
````

````python
retail.groupby(['family', 'store_nbr'], as_index=False)[['sales']].sum() 
````

---

## LABORATORY

````python
premier_league.head()
````

````python
premier_league.groupby(['HomeTeam', 'season'])[['HomeGoals']].sum()
````

````python
premier_league.groupby(['HomeTeam', 'season'], as_index=False)['HomeGoals'].sum()
````

````python
premier_league.groupby(['season', 'HomeTeam'], sort=False)[['HomeGoals']].sum().sort_values(['season', 'HomeGoals'], ascending=[True, False])
````

````python
premier_league.groupby(['season', 'HomeTeam'], as_index=False)[['HomeGoals']].sum().query("HomeTeam == 'Arsenal'")
````
